```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image
```

```python
def load_image(image_path):
    image = Image.open(image_path).convert('RGB')
    transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])
    return transform(image).unsqueeze(0)  # shape: [1, C, H, W]
```

```python
def forward_diffusion(x0, timesteps=1000, beta_start=1e-4, beta_end=0.02):
    betas = torch.linspace(beta_start, beta_end, timesteps)
    alphas = 1.0 - betas
    alpha_cumprod = torch.cumprod(alphas, dim=0)

    x_t = x0.clone()
    for t in range(timesteps):
        noise = torch.randn_like(x0)
        x_t = torch.sqrt(alpha_cumprod[t]) * x0 + torch.sqrt(1 - alpha_cumprod[t]) * noise
    return x_t
```

```python
def reverse_diffusion(x_t, timesteps=1000, beta_start=1e-4, beta_end=0.02):
    betas = torch.linspace(beta_start, beta_end, timesteps)
    alphas = 1.0 - betas
    alpha_cumprod = torch.cumprod(alphas, dim=0)

    for t in reversed(range(timesteps)):
        noise = torch.randn_like(x_t)
        x_t = (x_t - torch.sqrt(1 - alpha_cumprod[t]) * noise) / torch.sqrt(alpha_cumprod[t])
    return x_t
```

```python
def show_image(tensor_img, title="Image"):
    img = tensor_img.clone().detach().squeeze(0)
    img = img.permute(1, 2, 0)  # [C, H, W] -> [H, W, C]
    img = (img + 1) / 2  # Denormalize to [0, 1]
    img = torch.clamp(img, 0, 1)
    plt.imshow(img.cpu().numpy())
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image_path = '/content/download.jpeg'  # Change this to your image file
x0 = load_image(image_path)
show_image(x0, "Original Image")
```

```python
noised_img = forward_diffusion(x0)
show_image(noised_img, "Noised Image (Forward Diffusion)")

recovered_img = reverse_diffusion(noised_img)
show_image(recovered_img, "Recovered Image (Reverse Diffusion)")
```
